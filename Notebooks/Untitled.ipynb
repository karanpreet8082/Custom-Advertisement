```python

image = c.copy()
height, width, channels = image.shape

# Define the new dimensions for the magnified image (2X)
new_height = height * 2
new_width = width * 2

# Resize the image to the new dimensions using the nearest-neighbor interpolation
magnified_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_NEAREST)
```
